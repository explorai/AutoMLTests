```python
from sklearn.datasets import make_blobs, make_moons, make_circles, make_regression, make_checkerboard
from matplotlib import pyplot as plt
from pandas import DataFrame
# https://machinelearningmastery.com/generate-test-datasets-python-scikit-learn/
# generate 2d classification dataset
X, y = make_blobs(n_samples=100, centers=4, n_features=2)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green', 3:'yellow'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
# generate 2d classification dataset
X2, y2 = make_moons(n_samples=100, noise=0.15)
# scatter plot, dots colored by class value
df2 = DataFrame(dict(x=X2[:,0], y=X2[:,1], label=y2))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df2.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
# generate 2d classification dataset
X3, y3 = make_circles(n_samples=100, noise=0.075)
# scatter plot, dots colored by class value
df3 = DataFrame(dict(x=X3[:,0], y=X3[:,1], label=y3))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df3.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
# generate regression dataset
X4, y4 = make_regression(n_samples=100, n_features=1, noise=15)
# plot regression dataset
df4 = DataFrame(dict(x=X4[:,0], y=y4))
plt.scatter(X4,y4)
plt.show()
```

```python
df.to_csv('datasets/blobs.csv', index=False)
df2.to_csv('datasets/moons.csv', index=False)
df3.to_csv('datasets/circles.csv', index=False)
df4.to_csv('datasets/regression.csv', index=False)
```
