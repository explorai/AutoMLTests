```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
from utils import load
blobs_data_train = h2o.import_file("datasets/blobs_train.csv")
blobs_data_test = h2o.import_file("datasets/blobs_test.csv")
blobs_data_train['label'] = blobs_data_train['label'].asfactor()
blobs_data_test['label'] = blobs_data_test['label'].asfactor()
blobs_predictor = H2OAutoML(max_models=10, seed=1)
blobs_predictor.train(y='label', training_frame=blobs_data_train)
```

```python
blobs_predictor.leader.model_performance(blobs_data_test)
blobs_predictor.leader.explain(blobs_data_test)
```
